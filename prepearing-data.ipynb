```python
data=pd.DataFrame()
```

```python
for stock in bist100_stocks:
    df = yf.download(stock, start="2010-01-01", end="2025-01-01")  # Kapanış fiyatlarını al
    data[stock] = df["Close"]  # Düzeltilmiş kapanış fiyatlarını sakla
```

```python
data.head()
```

```python
data.to_csv("bist100_closed_price_data.csv")
```

```python
data.isnull()
```

```python
returns = data.pct_change()
```

```python
returns.tail()
returns1=returns.dropna()
returns1.shape
```

```python
returns.to_csv("bist100_daily_pricechange.csv")
```

```python
expected_returns = returns.mean()
```

```python
expected_returns.head()
expected_returns_up_to_down=expected_returns.sort_values(ascending=False)
expected_returns_up_to_down.to_csv("bist100_expected_returns_ranking.csv")
```

```python
std_returns=returns.std()
```

```python
std_returns.head()
```

```python
std_returns_up_to_down=std_returns.sort_values(ascending=False)
std_returns_up_to_down.to_csv("bist100_risky_ranking.csv")
```

```python
import matplotlib.pyplot as plt
# Günlük volatiliteyi grafikle gösterme
plt.figure(figsize=(50,15))
std_returns_up_to_down.plot(kind='bar', color='skyblue')
plt.title("Hisselerin Günlük Volatilitesi (Büyükten Küçüğe)")
plt.xlabel("Hisseler")
plt.ylabel("Günlük Volatilite")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
